```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import os

# Get path to home directory
home_dir = os.path.expanduser("~")

# Create database file in home directory
db_file = os.path.join(home_dir, 'student_grades.db')

# Connect to SQLite database (creates it if it doesn't exist)
conn = sqlite3.connect(db_file)
cursor = conn.cursor()

# Create students table
cursor.execute('''
CREATE TABLE IF NOT EXISTS students (
    student_id INTEGER PRIMARY KEY,
    first_name TEXT,
    last_name TEXT
)
''')

# Create grades table
cursor.execute('''
CREATE TABLE IF NOT EXISTS grades (
    grade_id INTEGER PRIMARY KEY,
    student_id INTEGER,
    subject TEXT,
    grade INTEGER,
    FOREIGN KEY (student_id) REFERENCES students (student_id)
)
''')

# Insert students data
students = [
    (1, 'Alice', 'Johnson'),
    (2, 'Bob', 'Smith'),
    (3, 'Carol', 'White'),
    (4, 'David', 'Brown'),
    (5, 'Eve', 'Davis')
]
cursor.executemany('INSERT OR REPLACE INTO students (student_id, first_name, last_name) VALUES (?, ?, ?)', students)

# Insert grades data
grades = [
    (1, 1, 'Math', 95), (2, 1, 'English', 88), (3, 1, 'History', 90),
    (4, 2, 'Math', 82), (5, 2, 'English', 76), (6, 2, 'History', 85),
    (7, 3, 'Math', 91), (8, 3, 'English', 93), (9, 3, 'History', 88),
    (10, 4, 'Math', 78), (11, 4, 'English', 85), (12, 4, 'History', 92),
    (13, 5, 'Math', 87), (14, 5, 'English', 89), (15, 5, 'History', 94)
]
cursor.executemany('INSERT OR REPLACE INTO grades (grade_id, student_id, subject, grade) VALUES (?, ?, ?, ?)', grades)

# Commit changes
conn.commit()

print(f"Database created at: {db_file}")

# Function to execute SQL queries and return results as a DataFrame
def execute_query(query):
    return pd.read_sql_query(query, conn)

# 1. Retrieve all students' names and their grades
query1 = '''
SELECT s.student_id, s.first_name, s.last_name, g.grade_id, g.subject, g.grade
FROM students s
JOIN grades g ON s.student_id = g.student_id
ORDER BY s.student_id, g.grade_id
'''
df1 = execute_query(query1)
print("1. All students' names and grades:")
print(df1)
print()

# 2. Find the average grade for each student
query2 = '''
SELECT s.student_id, s.first_name, s.last_name, AVG(g.grade) as avg_grade
FROM students s
JOIN grades g ON s.student_id = g.student_id
GROUP BY s.student_id
ORDER BY s.student_id
'''
df2 = execute_query(query2)
print("2. Average grade for each student:")
print(df2)
print()

# 3. Find the student with the highest average grade
query3 = '''
SELECT s.student_id, s.first_name, s.last_name, AVG(g.grade) as avg_grade
FROM students s
JOIN grades g ON s.student_id = g.student_id
GROUP BY s.student_id
ORDER BY avg_grade DESC
LIMIT 1
'''
df3 = execute_query(query3)
print("3. Student with the highest average grade:")
print(df3)
print()

# 4. Find the average grade for the Math subject
query4 = '''
SELECT AVG(grade) as avg_math_grade
FROM grades
WHERE subject = 'Math'
'''
df4 = execute_query(query4)
print("4. Average grade for Math:")
print(df4)
print()

# 5. List all students who scored above 90 in any subject
query5 = '''
SELECT DISTINCT s.student_id, s.first_name, s.last_name
FROM students s
JOIN grades g ON s.student_id = g.student_id
WHERE g.grade > 90
ORDER BY s.student_id
'''
df5 = execute_query(query5)
print("5. Students who scored above 90 in any subject:")
print(df5)
print()

# Bonus: Find the student with the highest grade in each subject
bonus_query = '''
SELECT s.student_id, s.first_name, s.last_name, g.subject, g.grade
FROM students s
JOIN grades g ON s.student_id = g.student_id
WHERE (g.subject, g.grade) IN (
    SELECT subject, MAX(grade)
    FROM grades
    GROUP BY subject
)
ORDER BY g.subject
'''
df_bonus = execute_query(bonus_query)
print("Bonus: Student with the highest grade in each subject:")
print(df_bonus)
print()

# Visualizations remain the same as in your original code
# ...

# Close the database connection
conn.close()
```

```python
conn = sqlite3.connect('student_grades.db')

# Query to get average grade for each student
query = '''
SELECT s.first_name, s.last_name, AVG(g.grade) as avg_grade
FROM students s
JOIN grades g ON s.student_id = g.student_id
GROUP BY s.student_id
'''

# Load query results into a DataFrame
df = pd.read_sql_query(query, conn)

# Visualize average grades
plt.figure(figsize=(10, 6))
plt.bar(df['first_name'] + ' ' + df['last_name'], df['avg_grade'])
plt.title('Average Grades by Student')
plt.xlabel('Student')
plt.ylabel('Average Grade')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

conn.close()
```
